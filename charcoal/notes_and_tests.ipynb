### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
```

### Opening image files

```python
# Declares variables and anonymous functions
upfr_dataset_path = "../dataset_upfr/"
retrieve_tif = lambda x: x[-4:]==".tif"

# Retrieves .tif images
image_files = list(filter(retrieve_tif, sorted(os.listdir(upfr_dataset_path))))

# Renames images from .tif to .tiff

for f in image_files:
    old_name = f"{upfr_dataset_path}{f}"
    new_name = f"{upfr_dataset_path}{f}f"
    os.rename(old_name, new_name)
```

```python
with Image.open(upfr_dataset_path+image_files[0]) as img:
    i = np.asarray(img, dtype = np.uint8)
```

```python
plt.figure(figsize=(15,9))
plt.imshow(i, cmap="gray")
plt.title(image_files[0])
plt.show()
```
